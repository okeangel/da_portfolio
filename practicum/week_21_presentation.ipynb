```python
def load(name, ext='csv', sep=','):
    """Return pandas DataFrame by filename"""
    filename = name + '.' + ext
    try:
        df = pd.read_csv(filename)
    except FileNotFoundError:
        df = pd.read_csv('https://code.s3.yandex.net/datasets/' + filename, sep=sep)
    correct_column_names(df)
    overview(df)
    return df
```

```python
def overview(data):
    """Display an overview of the data to detect common data problems"""
    display(data.sample(5))
    dupes = data.duplicated().sum()
    dupes_rate = dupes / len(data)
    print(f'Полностью совпадающих строк: {dupes} ({dupes_rate:.2%})\n')
    data.info()
    display(count_nans(data))
    print(get_filled_rows_share(data))
```

```python
def count_nans(data):
    """Return a pandas DataFrame with the number of missing values"""
    nans = []
    for column in data.columns:
        # count the number of missing values
        missing = data[column].isna().sum()
        # calculate the share of missing values
        part = round(missing * 100 / len(data), 2)
        # count the number of distinct values
        uniques = data[column].sort_values().nunique()
        nans.append([column, missing, part, uniques])

    return pd.DataFrame(
        data=nans, columns=['column', 'na_count', 'na_%', 'unique_count']
    ).sort_values(['na_count', 'unique_count'], ascending=False).set_index('column')
```

```python
def get_filled_rows_share(df):
    """Return a string containing number of rows with no missing values"""
    share = 1 - df.isna().any(axis=1).sum() / len(df)
    return f'Заполненных строк без пропусков: {share:.02%}'
```

##### Исправление данных

```python
def correct_column_names(df):
    """Convert column names to valid snake_case in place"""
    df.columns = df.columns.str.strip()
    df.columns = list(map(to_snake_case, df.columns))

def to_snake_case(name):
    """Return name converted to valid snake_case"""    
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    name = re.sub('__([A-Z])', r'_\1', name)
    name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
    return name.lower()
```

```python
def fillnaby(df, column, groupby, func='median'):
    """Fill in missing values with group-aggregated values in-place"""
    df[column] = df[column].fillna(
        df.groupby(groupby)[column].transform(func))
```

### Преодоление неявных и явных дубликатов

Бытовая привычка говорит нам, что если название двух объектов одинаково или слишком похоже - то это одинаковые объекты, например, представители одной сети. Юристы и другие специалисты могут напомнить, что это не всегда так. Вполне могу существовать два предприятия с одинаковыми названиями, никак не связанные между собой. И степень связанности может быть разной - общая администрация, общее юрлицо, общий владелец, общая франшиза, стихийное использование общего громкого названия (особенно если оно не защищено юридически). Поэтому сложно сказать, что именно мы будем считать сетью. Опираться только на названия - не очень надёжно.

Также учитываем, что названия могут быть написаны по-разному просто по стечению обстоятельств. Например, названия предприятий Яндекса записаны в разных источниках по разному, то с точкой, то без неё. Названия записаны по-разному, но предприятия Яндекса определённо не самостоятельны, а объединены.

Основной смысл несут буквы и числа. Именно они произносятся, именно они составляют речь. Пробелы, знаки препинания, служебные символы, регистр - важны гораздо меньше, и именно они обычно различаются при написании названия одного и того же объекта, поскольку в речи они трудно различимы.

Поэтому уникальность строки будем определять не по всему её содержимому, а только оставив в строке буквы и цифры, без всего, без пробелов. Это тоже вполне читаемый текст, это и есть суть текста, суть названия.

Если строка содержит те же буквы и цифры в том же порядке, как другая строка - они считаются одинаковыми. Сперва поищем, сколько одинаковых по смыслу строк в текстовых столбцах.

```python
def get_meaning_pattern(text):
    # имеющими значение будем считать только буквы и цифры
    return ''.join(char for char in text.lower() if char.isalpha() or char.isdigit())

# проверяем все текстовые поля,
# есть ли в них строки, различающиеся не смыслом, а формой записи
for column in places.select_dtypes(include='object').columns:
    values = places[column].dropna().unique()
    patterns = set(map(get_meaning_pattern, values))
    if len(patterns) != len(values):
        print('Столбец {} содержит неодинаковое количество различных значений:\n'
              'по смыслу их {} разных, по форме записи {} разных'.format(
                  column, len(patterns), len(values)))
        print(values + '\n')
```

Подготовим функцию, которая, получив строку таблицы, возвращает текст из выбранной колонки. Если ранее уже встречалась другая формулировка этого же текста - то вернёт ту, ранее встречавшуюся. Смыслы и соответствующие им формулировки накапливаются во внешнем словаре. Поэтому перед использованием функции словарь надо создавать (пустым).

```python
def get_mean_text(row, column):
    pattern = get_meaning_pattern(row[column])    
    if pattern in patterns.keys() and row[column] != patterns[pattern]:
        print(f"{row[column]} -> {patterns[pattern]}")
        return patterns[pattern]
    else:
        patterns[pattern] = row[column]
        return row[column]
```

Проверка на уникальность показала, что 'names' и 'addresses' содержат неявные дубликаты. Поскольку дальнейшие исследования будут использовать названия - то исходные столбцы менять не будем, а добавим дополнительные 2 столбца, с "подразумеваемыми" названиями и адресами.

```python
patterns = dict()
column = 'name'
places[f'mean_{column}'] = places.apply(
    lambda row: get_mean_text(row, column), axis=1)
```

### Добавляем `street`

... чтобы проанализировать, на каких улицах есть общепит.

```python
street_attributes = [
    'улица', 'ул.', 'просп', 'проезд', 'бул.', 'бульвар', 'б-р',
    'шоссе', 'площ',
]

def get_street(place):
    addr = place['address'].split(', ')

    # иногда в адресах вторым идёт какой-нибудь район или посёлок,
    # а в конце идёт не только дом, но и строение,
    # поэтому будем искать улицу не по позиции в адресе,
    # а очевидным способом - по упоминанию слова "улица" или аналогичного
    for part in addr:
        for attr in street_attributes:
            if attr in part:
                return part

    # если не найдём - вернём второй элемент адреса
    return addr[1]

places['street'] = places.apply(get_street, axis=1)
places.sample(10)[['address', 'street']]
```

... чтобы проверить, выделяются ли рестораны, работающие без перерывов, чем-нибудь особенным.
