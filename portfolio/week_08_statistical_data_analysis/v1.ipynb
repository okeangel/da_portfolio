#### Тарифы

##### Общие правила

- Каждый звонок округляется отдельно.
- Для интернет-трафика отдельные сессии не считаются.
- Секунды округляются "вверх" до минут, мегабайты — "вверх" до гигабайт.
- Согласно заданию, считаем, что неиспользованные за месяц услуги не переносятся.

##### Тариф «Смарт»

- Ежемесячная плата: 550 рублей.
- Включено: 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика.
- Стоимость услуг сверх тарифного пакета:
  - минута разговора: 3 рубля;
  - сообщение: 3 рубля;
  - 1 Гб интернет-трафика: 200 рублей.

##### Тариф «Ультра»

- Ежемесячная плата: 1950 рублей.
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика.
- Стоимость услуг сверх тарифного пакета:
  - минута разговора: 1 рубль;
  - сообщение: 1 рубль;
  - 1 Гб интернет-трафика: 150 рублей.

#### Описание выборки

##### Таблица tariffs (информация о тарифах):

- tariff_name — название тарифа
- rub_monthly_fee — ежемесячная абонентская плата в рублях
- minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
- messages_included — количество сообщений в месяц, включённых в абонентскую плату
- mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

##### Таблица users (информация о пользователях):

- user_id — уникальный идентификатор пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя
- age — возраст пользователя (годы)
- reg_date — дата подключения тарифа (день, месяц, год)
- churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- city — город проживания пользователя
- tariff — название тарифного плана

##### Таблица calls (информация о звонках):

- id — уникальный номер звонка
- call_date — дата звонка
- duration — длительность звонка в минутах (звонок с продолжительностью 0 - это звонок без ответа)
- user_id — идентификатор пользователя, сделавшего звонок

##### Таблица messages (информация о сообщениях):

- id — уникальный номер сообщения
- message_date — дата сообщения
- user_id — идентификатор пользователя, отправившего сообщение

##### Таблица internet (информация об интернет-сессиях):

- id — уникальный номер сессии
- mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
- session_date — дата интернет-сессии
- user_id — идентификатор пользователя

```python
def get_nans(data):
    nans = []
    for column in data.columns:
        missing = data[column].isna().sum()
        part = missing / len(data)
        uniques = data[column].sort_values().unique()
        nans.append([column, part, len(uniques)])

    return pd.DataFrame(
        data=nans, columns=['column', 'na_part', 'unique_count']
    ).sort_values(['na_part', 'unique_count'], ascending=False).set_index('column')
```

```python
FIGSIZE = (10, 5)

# формирует и выводит график, подстроенный под наши нужды
def histf(column, data, show=True, bins=100, qtl=1, start=None,
          stop=None, figsize=FIGSIZE, **kwargs):
    fixed = {
        'y': column,
        'kind': 'hist',
        'bins': min(bins, len(data[column].unique())),
        'figsize': figsize,
        'title': column,
        'legend': False,
    }
    if 'range' not in kwargs.keys():
        if not start:
            start = data[column].min()
        if not stop:
            stop = data[column].quantile(qtl)
        kwargs['range'] = (start, stop)
    data.plot(**fixed, **kwargs)
    if show:
        plt.show()

# два графика: все случаи, и те, что менее указанного квантиля
def dhistf(column, data, qtl=0.975, figsize=FIGSIZE, **kwargs):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    fig.suptitle('This is a somewhat long figure title', fontsize=16)
    
    ax1.set_title('full_data')
    ax2.set_title(f'{qtl:.2%} quantile')
    histf(column, data=data, ax=ax1, subplots=True, show=False, **kwargs)
    histf(column, data=data, ax=ax2, subplots=True, show=False, qtl=qtl, **kwargs)
    plt.show()
# Развитая форма этого чудовища должна принимать список квантилей,
# и строить столько графиков, сколько квантилей указано,
# и качестве центра диапазона выбирать медиану.
# Пора перейти к аналитике.
```

```python
def typical(data, column, ratio=1.5):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    whiskers = iqr * ratio
    low = q1 - whiskers
    high = q3 + whiskers
    
    print(f'{column.title()}: min = {data[column].min()}, max = {data[column].max()}, normal range = [{low:.2f} ... {high:.2f}]')
    return data.query(f'@low <= {column} <= @high')
```

## Расходы каждого пользователя в месяц

### Округляем минуты "вверх"

Наша задача - сравнить расходы пользователей двух тарифов. Перед нами стоит задача узнать, какой тариф приносит больше выручки. Таким образом, нас интересует учёт финансовых расходов пользователя (сколько потратил рублей), а не технических (сколько потратил ресурсов).

Продолжительность звонка учитывается двумя способами:
- хранится в виде неполных минут (то что фактически употребил);
- тарифицирутся как полные минуты (то, за что заплатил).

Для финансового учёта минуты по каждому звонку нужно округлить "вверх".

```python
calls['duration_ceiled'] = calls['duration'].apply(np.ceil)
display(calls)
```

### Определяем минимальный размер периодов, которые можно сравнивать

Денежные списания с баланса пользователя имеют периодичность, заданную правилами тарифа: определённая сумма (ежемесячная плата) списывается с баланса раз в месяц. Такая периодичность вызвана поведением оператора, а не поведением пользователей. Поведение оператора нас не интересует. Чтобы эта цикличность не приковывала к себе внимание, объединим расходы в группы по месяцам. Цикличность оператора будет скрыта, на виду останется поведение пользователей.

### Определяем период, за который у нас есть данные

Чтобы определить, какой диапазон месяцев использовать, найдём самую раннюю и самую позднюю дату среди данных.

```python
date_columns = []
for table in [calls, messages, internet, users]:
    for column in table.columns:
        if '_date' in column:
            date_columns.append((table, column))
            print(table[column].describe(datetime_is_numeric=True))
```

Все данные укладываются в диапазон между 1 января и 31 декабря одного и того же 2018 года. Тогда их можно группировать просто по номерам месяцев.

```python
for table in [calls, messages, internet, users]:
    for column in table.columns:
        if column.endswith('_date'):
            new_column = column[:-5] + '_month'
            table[new_column] = table[column].dt.month
    display(table)
```

Обратим внимание, что те месяцы, когда пользователь зарегистрировался и прекратил регистрацию - он не использовал связь полный месяц. Неполные месяцы не следует добавлять в статистику полных месяцев (по крайнем мере, без предварительной подготовки).

Также, задание не сообщает нам, по каким правилам взимается плата за месяц: в одну и ту же дату календарного месяца, или в зависимости от даты регистрации пользователя. Если в одну и ту же - то уменьшается ли оплата за неполный месяц, или взимается как за полный.

Чтобы эта неопределённость не влияла на результат, мы не будем учитывать неполные месяцы.

### Создаём таблицу с расходами пользователей за каждый период (месяц)

```python
calls_by_user_and_month = calls.pivot_table(
    index=['user_id', 'call_month'],
    values=['duration_ceiled'], aggfunc=['count', 'sum'])
calls_by_user_and_month.columns = ['call_count', 'call_duration']

messages_by_user_and_month = messages.pivot_table(
    index=['user_id', 'message_month'],
    values=['id'], aggfunc=['count'])
messages_by_user_and_month.columns = ['message_count']

internet_by_user_and_month = internet.pivot_table(
    index=['user_id', 'session_month'],
    values=['mb_used'], aggfunc=['count', 'sum'])
internet_by_user_and_month.loc[:, ('sum', 'mb_used')] /= 1024
internet_by_user_and_month.columns = ['internet_session_count', 'internet_gb']
internet_by_user_and_month['internet_gb'] = internet_by_user_and_month['internet_gb'].apply(
    np.ceil
)

consumption = calls_by_user_and_month.join(
    [messages_by_user_and_month, internet_by_user_and_month],
    how='outer',
).fillna(0).reset_index().rename(columns={'level_1': 'month'}).astype('int64')

display(consumption)
```

Добавим в таблицу название тарифа.

```python
consumption['tariff'] = consumption.merge(
    users,
    on='user_id',
    how='left',
)['tariff']

consumption = consumption.merge(
    tariffs, how='left', left_on='tariff', right_on='tariff_name')
display(consumption)
```

Добавим траты на потребление.

```python
def calc_expenses(row):
    """Возвращает сумму денежных расходов согласно правилам тарифа:
    - оплата за неиспользованные ресурсы из лимита - не вычитается из суммы;
    - оплата за ресурсы сверх лимита - прибавляется к сумме.
    """
    return (row['rub_monthly_fee']
            + max(row['call_duration'] - row['minutes_included'], 0)
            * row['rub_per_minute']
            + max(row['message_count'] - row['messages_included'], 0)
            * row['rub_per_message']
            + max(row['internet_session_count'] - row['mb_per_month_included'] / 1024, 0)
            * row['rub_per_gb'])

consumption['expenses'] = consumption.apply(calc_expenses, axis=1)
display(consumption)
```

Отбросим все неполные месяцы. Останутся только полные месяцы. Это уменьшает итоговое количество данных. Он это делает данные не зависимыми от даты подключения внутри месяца. Они будут яснее показывать поведение пользователя именно в течение типичного полного месяца.

```python
consumption[['reg_month', 'churn_month']] = consumption.merge(
    users,
    on='user_id',
    how='left',
)[['reg_month', 'churn_month']]

display(consumption[['month', 'reg_month', 'churn_month']])
```

```python
consumption = consumption[
    (consumption['month'] != consumption['reg_month'])
    & (consumption['month'] != consumption['churn_month'])
].reset_index(drop=True)

display(consumption)
```

Оставим только нужные колонки.

```python
consumption = consumption.drop(
    labels=(list(tariffs.columns) + ['reg_month', 'churn_month']),
    axis=1)
display(consumption)
```

- включая сумму оплат за услуги;
- исключая неполные месяцы (те, в которые клиент подключился или отключился).
