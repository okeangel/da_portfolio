## Описание проекта

### Заказчик
Федеральный оператор сотовой связи "Мегалайн", коммерческий департамент.

### Задача

Определить, какой из тарифов приносит "больше денег". На основе анализа будет "скорректирован" рекламный бюджет. 

### Входные данные

- Описания тарифов.
- Выборка информации о тарифицируемых действиях 500 клиентов "Мегалайн" за 2018 год.

#### Тарифы

##### Общие правила

- Каждый звонок округляется отдельно.
- Для интернет-трафика отдельные сессии не считаются.
- Секунды округляются "вверх" до минут, мегабайты — "вверх" до гигабайт.
- Согласно заданию, считаем, что неиспользованные за месяц услуги не переносятся.

##### Тариф «Смарт»

- Ежемесячная плата: 550 рублей.
- Включено: 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика.
- Стоимость услуг сверх тарифного пакета:
  - минута разговора: 3 рубля;
  - сообщение: 3 рубля;
  - 1 Гб интернет-трафика: 200 рублей.

##### Тариф «Ультра»

- Ежемесячная плата: 1950 рублей.
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика.
- Стоимость услуг сверх тарифного пакета:
  - минута разговора: 1 рубль;
  - сообщение: 1 рубль;
  - 1 Гб интернет-трафика: 150 рублей.

#### Описание выборки

##### Таблица tariffs (информация о тарифах):

- tariff_name — название тарифа
- rub_monthly_fee — ежемесячная абонентская плата в рублях
- minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
- messages_included — количество сообщений в месяц, включённых в абонентскую плату
- mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

##### Таблица users (информация о пользователях):

- user_id — уникальный идентификатор пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя
- age — возраст пользователя (годы)
- reg_date — дата подключения тарифа (день, месяц, год)
- churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- city — город проживания пользователя
- tariff — название тарифного плана

##### Таблица calls (информация о звонках):

- id — уникальный номер звонка
- call_date — дата звонка
- duration — длительность звонка в минутах (звонок с продолжительностью 0 - это звонок без ответа)
- user_id — идентификатор пользователя, сделавшего звонок

##### Таблица messages (информация о сообщениях):

- id — уникальный номер сообщения
- message_date — дата сообщения
- user_id — идентификатор пользователя, отправившего сообщение

##### Таблица internet (информация об интернет-сессиях):

- id — уникальный номер сессии
- mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
- session_date — дата интернет-сессии
- user_id — идентификатор пользователя

## Гипотезы и проверки

###  Средняя выручка пользователей тарифов различаются

Выручкой оператора связи становятся те финансовые расходы, которые пользователь отнёс на связь. Главный вопрос, поставленный оператором связи - пользователи какого тарифа приносят больше выручки?

В нашем распоряжении выборка клиентов, разделённая на две части. Поскольку это лишь две выборки, сравнивать напрямую средние величины выборок будет неверным. Нам поставлен вопрос не о выборках, а о совокупности всех клиентов, имеющихся и потенциальных.

Используем T-тест - он позволяет оценить, насколько высока вероятность, что наши выборки принадлежат одной генеральной совокупности. Насколько вероятно, что их непохожесть - это случайность.

По умолчанию принято формулировать гипотезу так, чтобы она указывала на равенство величин. В нашем случае - на равенство средней выручки с каждого тарифа.

Таким образом, прямая гипотеза H0:
- средние величины расходов `expenses` пользователей обоих тарифов будут равны.

Альтернативная гипотеза H1:
- средние величины расходов `expenses` у пользователей обоих тарифов различаются.

Допустимый уровень значимости (вероятность того, что мы ошибочно отвергнем гипотезу H0) возьмём 0.001 как одно из общепринятых значений. У нас нет статистики, как часто не срабатывала данная гипотеза в таких случаях. Однако психологически комфортно считать, что тест подведёт нас всего один раз в несколько лет, если использовать его каждый день.

```python
_, pvalue = st.ttest_ind(
    smart_consumption['expenses'],
    ultra_consumption['expenses'],
    equal_var=False,
)
print('P-value:', pvalue)
alpha = 0.001
if pvalue < alpha:
    print('H0 не подтверждается. Средние величины расходов не равны.')
else:
    print('H0 не опровергнута. Средние величины расходов можно считать равными.')
```

**Итоги:**
- Есть статистическое основание утверждать, что в целом выручка от пользователей тарифов "Smart" и "Ultra" различается.
- Ранее мы установили, что в выборке наблюдается факт: денежные расходы пользователей тарифа "Ultra" на треть выше по сравнению со "Smart".
- Опираясь на эти два факта, делаем заключение: тариф "Ultra" приносит на треть больше выручки с одного пользователя, чем "Smart".

### Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

По общему мнению, в Москве люди расходуют больше. Предполагаем, что коммерческий отдел ожидает, что в Москве расходы на связь выше, чем за её пределами. Мы проверим эту гипотезу на нашей выборке.

Формулируем гипотезы.

- H0: Средняя выручка пользователей из Москвы равна средней выручке пользователей не из Москвы.

- H1: Средняя выручка пользователей из Москвы больше чем средняя выручка пользователей не из Москвы.

Уровень значимости установим как и в предыдущем тесте. Также будем учитывать, что нас интересует отклонение в одну сторону (Москва больше регионов), поэтому p-value будет уменьшен вдвое, и мы посчитаем две средних величины выборки: для Москвичей и для остальных пользователей.

Разделим выборку на две.

```python
moscow_user_ids = users.query('city == "Москва"')['user_id']

moscow_consumption = consumption.query('user_id in @moscow_user_ids')
regions_consumption = consumption.query('user_id not in @moscow_user_ids')

print('Выборка разделена успешно.' if len(moscow_consumption) + len(regions_consumption)
      == len(consumption) else 'Выборка разделена не правильно.')
```

```python
moscow_mean = moscow_consumption['expenses'].mean() 
moscow_std = moscow_consumption['expenses'].std() 
regions_mean = regions_consumption['expenses'].mean()
regions_std = regions_consumption['expenses'].std()
print("Средняя выручка с пользователя в месяц:")
print(f'Москва - {moscow_mean:.02f} (std = {moscow_std:.02f}),')
print(f'регионы - {regions_mean:.02f} (std = {regions_std:.02f}).')

_, pvalue = st.ttest_ind(
    moscow_consumption['expenses'],
    regions_consumption['expenses'],
    equal_var=False)

print('P-value:', pvalue)
if (pvalue / 2 < alpha) and (moscow_mean > regions_mean):
    print('H0 не подтверждается. Средняя выручка в Москве выше, чем в регионах.')
else:
    print('H0 не опровергнута. Нельзя считать, что в Москве средняя выручка статистически выше.')
```

**Итоги:**
- Опираясь на результат статистического T-теста, можно утверждать, что не обнаружена разница в выручке между пользователями оператора в Москве и за её пределами.
