## Описание проекта

### Задача

Установить признаки, при наличии которых объявление можно считать аномальным. Это позволит автоматически выявлять потенциально мошеннические объявления.

### Входные данные

Выборка из объявлений Яндекс.Недвижимости о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

### Инструменты

Используем [Anaconda](https://www.anaconda.com/), окружение `ds_da_practicum_env.yml` "для Mac OS" (от 2022-03-23).

### План работы

- Знакомимся с данными.
- Проверяем пропуски, заполняем.
  - Определите и изучите пропущенные значения:
    - Ищем логичные решения, чем заполнить пропуски;
    - Если решение есть - заполняем;
    - Описываем, почему именно в этих столбцах и как выбрали значения.
    - Описываем, какие причины могли привести к пропускам в данных.
  - Приводим данные к нужным типам:
    - Определяем, в каких столбцах.
    - Почему меняем?
- Добавляем данные:
  - цену квадратного метра;
  - день недели, месяц и год публикации объявления;
  - этаж квартиры; варианты — первый, последний, другой;
  - соотношение жилой и общей площади, а также отношение площади кухни к общей.
- Изучаем данные:
  - Площадь, цена, число комнат, высота потолков (гистограммы).
  - Время продажи квартиры (гистограмма, среднее и медиану).
    - Сколько обычно занимает продажа?
    - Когда можно считать, что продажи прошли очень быстро, а когда необычно долго?

####  Пропуски до обработки

```python
def get_nans(data):
    nans = []
    for column in data.columns:
        missing = data[column].isna().sum()
        part = missing / len(data)
        uniques = data[column].sort_values().unique()
        nans.append([column, part, len(uniques)])

    return pd.DataFrame(
        data=nans, columns=['column', 'na_part', 'unique_count']
    ).sort_values(['na_part', 'unique_count'], ascending=False).set_index('column')

nans_before = get_nans(realty)
display(nans_before)
```

Пропусков много. Может, использовать только строки без пропусков? Сколько их?

```python
def print_filled_rows_share(df):
    share = 1 - df.isna().any(axis=1).sum() / len(df)
    print(f'Заполненных строк без пропусков: {share:.02%}')

print_filled_rows_share(realty)
```

Содержащих всю информацию строк всего 1%. Нужно заполнить недостающие данные. 

Взглянем на то, что содержат заполенные данные. Возможно, где-то в распределениях сияют дыры, в которые и нужно уложить незаполненные данные. Вверху графики наиболее редких данных, внизу - наиболее полных.

```python
def headline(headline, width=50):
    """Возвращает заголовок, набранный "простым текстом".
    """
    return f'\n{headline.upper():=^{width}}'
```

В описании к данным прямо сказано, что в выборке - только квартиры, нет особняков. Поэтому мы не ждём ситуации, когда "нет балконов потому, что есть три веранды (или крыльца)". В квартире подразумевается только один тип "наружней" площади: балкон. Чем больше квартира - тем больше балконов.

*(Особняком назовём любой жилой дом, в котором живёт одна, обособленная семья. Это короткое и очевидное слово. "Частный дом" звучит так, будто непременно находится в частной собственности, а "жилой дом" или "отдельный дом" - будто остальные дома - нежилые или "совмещённые".)*

Как проверить: построим распределения жилой площади по каждому из вариантов "количества балконов".

```python
def plot_hist(data, title, bins=49):
    data.plot(
        kind='hist',
        bins=bins,
        title=title,
        range=(10, 60)
    )
    plt.show()

plot_hist(realty[realty['balcony'].isna()]['living_area'], 'Балконов: не указано')
for balcony in realty['balcony'].dropna().sort_values().unique():
    plot_hist(
        realty[realty['balcony'] == balcony]['living_area'],
        f'Балконов: {balcony}'
    )
```

Распределение при неуказанном количестве балконов такое же, как если указано 0, 1 или 2 балкона. При количестве балконов более 2-х - распределение меняется, но таких случаев мало самих по себе.

Однако мы видим пики при определённых жилых площадах. Ярко выраженных три пика. Это могут быть пики жилых площадей 1-, 2- и 3- комнатных квартир.

Проследим распределение количества жилых комнат по каждому из значений количества балконов.

```python
def plot_hist(data, title, bins=9*2):
    data.plot(
        kind='hist',
        bins=bins,
        title=title,
        range=(0, 9)
    )
    plt.show()

plot_hist(realty[realty['balcony'].isna()]['rooms'], 'Балконов: не указано')
for balcony in realty['balcony'].dropna().sort_values().unique():
    plot_hist(
        realty[realty['balcony'] == balcony]['rooms'],
        f'Балконов: {balcony}'
    )
```

Квартиры без указания балконов похожи на квартиры с 0 балконов. Нексолько меньше - на квартиры с 1 и 2 балконами.

```python
def fillnaby(data, column, groupby, func='median'):
    data[column] = data[column].fillna(
        data.groupby(groupby)[column].transform(func))
```

Подготовим две функции. Одна рисует гистограммы, которые подстраиваются под данные. Вторая строит две такие гистограммы для сравнения.

```python
FIGSIZE = (10, 5)

# формирует и выводит график, подстроенный под наши нужды
def histf(column, data=realty, show=True, bins=100, qtl=1, start=None,
          stop=None, figsize=FIGSIZE, **kwargs):
    fixed = {
        'y': column,
        'kind': 'hist',
        'bins': min(bins, len(data[column].unique())),
        'figsize': figsize,
        'title': column,
        'legend': False,
    }
    if 'range' not in kwargs.keys():
        if not start:
            start = data[column].min()
        if not stop:
            stop = data[column].quantile(qtl)
        kwargs['range'] = (start, stop)
    data.plot(**fixed, **kwargs)
    if show:
        plt.show()

# два графика: все случаи, и те, что менее указанного квантиля
def dhistf(column, data=realty, qtl=0.975, figsize=FIGSIZE, **kwargs):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    fig.suptitle('This is a somewhat long figure title', fontsize=16)
    
    ax1.set_title('full_data')
    ax2.set_title(f'{qtl:.2%} quantile')
    histf(column, data=data, ax=ax1, subplots=True, show=False, **kwargs)
    histf(column, data=data, ax=ax2, subplots=True, show=False, qtl=qtl, **kwargs)
    plt.show()
# Развитая форма этого чудовища должна принимать список квантилей,
# и строить столько графиков, сколько квантилей указано,
# и качестве центра диапазона выбирать медиану.
# Пора перейти к аналитике.
```

### Убираем редкие и выбивающиеся значения

#### Убираем согласно правилу, автоматически

Мы обращаемся к изучению статистики тогда, когда признаём, что не вправе описаться на личный опыт, что личного опыта недостаточно. Что значит "редкие" или "выбивающиеся" значения? Любая граница, принятая "на основании опыта" - это граница, проведённая необоснованно. Мы выбрали путь - опираться на полученные данные и алгоритмы обработки данных. Это объдинённый опыт многих людей. Он надёжней личного опыта отдельно взятого аналитика.

Руководствуясь этой идеей, ицем подходящее правило: что считать "выбивающимся значением". Правило, которое изложено в курсе: "нормальными признаются значения между квартилей, редкими - те, которые отстоят от ближайшего квартиля не более чем на 1,5 межквартильных диапазона, выбросами - все оставшиеся." Изложим его в виде функции.

```python
def typical(data, column, ratio=1.5):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    whiskers = iqr * ratio
    low = q1 - whiskers
    high = q3 + whiskers
    
    print(f'{column.title()}: min = {data[column].min()}, max = {data[column].max()}, normal range = [{low:.2f} ... {high:.2f}]')
    return data.query(f'@low <= {column} <= @high')
```

```python
typical_realty = realty
main_properties = ['total_area', 'rooms', 'last_price_mln', 'ceiling_height']
print(headline('Обнаруженные и вычисленные величины'))
for property_column in main_properties:
    typical_realty = typical(typical_realty, property_column)
print(f'Типичной недвижимости: {len(typical_realty) / len(realty):.2%}')
print(headline('Графики даннных, очищенных от выбросов'))
for property_column in main_properties:
    histf(property_column, data=typical_realty)
```

Взглянув на графики, мы замечаем:
- стало удобно отображать типичные случаи;
- исчезли нетипичные, но реальные случаи:
  - квартиры с 7-ю комнатами - но ведь это значение не выглядит нереальным, пскольку существуют 6-комнатные;
  - потолки выше 3.12, которые действительно редки, но встречаются: потолки высотою по 5 метров на первых этажах домов, а в некоторых домах - и на каждом этаже.

**Итоги:**
- Фильтрование нужных результатов по IQR даёт результат, который в целом отбрасывает аномальные значения, оставляя типичные.
- Однако в выборку попала цена, близкая к нулевой. Её естественность под сомнением, посколку между нею и соседними значениями есть незаполенный диапазон. Говоря иначе, цены около 0 образуют локальную вершину, которую нужно рационально объяснить, либо признать аномалией.
- Этот метод хорош, когда:
  - нужно фильтровать автоматически:
    - когда много разнообразных колонок, датасетов;
    - когда аналитик не готов заниматься подготовкой данных, или аналитик отстутсвует;
  - когда аналитик не знает, что считать допустимым значением, а что - нереальным;
  - когда идёт изучение типичных случаев.
- Этот метод плох, когда:
  - нужно отделить экстремальные, но реальные случаи от нереальных.
- Теперь я лучше понимаю, почему рекомендательные сервисы рекомендуют то, что популярно, а не то, что меня интересует (а меня интересуют редкие, уникальные вещи) - возможно, я попадаю в категорию "статистические выбросы", и мой выбор не учитывается.

#### Полезные функции для отображения данных

Все собранные данные разделим на две группы: количественные (quantitative) и  категориальные (categorial).

Создадим три функции, которые позволят подстраивать графики, изменяя настройки в одном месте, а не в многих ячейках.

Количественные данные визуально будем изучать по диаграммам рассеивания, категориальные - по диаграммам-"ящикам".

Коэффициент корреляции может быть найден и для тех, и для других. Для удобства будем сортировать полученные величины начиная с наибольшей положительной корреляции.

Основные вопросы у нас - это связь итоговой цены квартиры с другими характеристиками. Поэтому по умолчанию по вертикали будет итоговая цена.


```python
quantitative_properties = [
    'last_price_mln', 'total_area', 'living_area_share', 'kitchen_area_share',
    'ceiling_height', 'city_centers_nearest', 'airports_nearest',
]

categorical_properties = [
    'rooms', 'balcony', 'floor_category',
    'is_apartment', 'is_studio', 'has_open_plan',
    'parks_around3000', 'ponds_around3000',
    'weekday', 'month', 'year',
]

def batch_corr(data, column='last_price_mln'):
    print(headline('Коэффициенты линейной корреляции'))
    display(
            data
            .reindex(columns=(quantitative_properties + categorical_properties))
            .corr()
            [column].sort_values(ascending=False)
    )

def batch_scatter(data, column='last_price_mln'):
    for prop in quantitative_properties:
        if prop == column:
            continue
        data.plot(
            kind='scatter', alpha=0.2, figsize=FIGSIZE,
            y=column,
            x=prop,
        )
    plt.show()

def batch_box(data, column='last_price_mln', by_props=categorical_properties):
    for prop in by_props:
        data.boxplot(
            column=column,
            by=prop,
            figsize=FIGSIZE,
        )
    plt.show()
```

#### Дата размещения

Информация о цене упоминает, что это цена на момент закрытия объявления, а не на момент подачи. Неявно подразумевается, что цену можно менять в любой день недели после подачи. Тогда не видно причин для корреляции между ценой, и днём подачи. Месяц и год подачи могут показать корреляцию, поскольку в среднем объявление висит менее года, и, таким образом, объявление не может быть изменено в любой случайный месяц, а может быть изменено лиш в месяц подачи, или в один из нескольких последующих.

Данные по дням могут быть полностью сглажены, данные по месяцам - иметь сглаженную, но ощутимую периодичность. Данные по годам можно считать почти не сглаженными.

```python
month_price = typical_realty.pivot_table(index='month',
                                         values='last_price_mln',
                                         aggfunc=['count', 'mean', 'median'])
month_price.columns = ['count', 'mean_price', 'median_price']
display(month_price)
```

Количество объявлений сильно зависит от месяца. Видна сезонность. Меньше всего в мае, далее количество объявлений нарастает, достигая пика в марте. А апреле предложение уменьшается, перед очередным маем. В декабре и январе - "провалы".

Предполагаем, что такое распределение - результат сложения двух явлений. Существует сезонная цикличность - с пиком зимой, и провалом летом. На неё накладываются "новогодние" и "майские" праздники, во время которых, как и в субботу-воскресение, объявлений публикуется меньше. Меньше рабочих дней в месяце - меньше объявлений. Это можно будет проверить точнее, если разложить объявления по неделям, прошедшим с начала года.

Нижний квартиль Q1 в течение всего года находится на одном уровне. Верхний квартиль Q3 заметно меняется по месяцам.

Не зная количества объявлений, мы могли бы предположить, что по месяцам меняется количество дорогих объявлений. Когда их больше - верхний квартиль поднимается. Однако, зная, что количество объявлений различается в 2-3 раза, мы понимаем, что это предположение было бы ошибочным.
